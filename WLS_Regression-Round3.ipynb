```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame

main_df = pd.read_csv('cancer_reg.csv')
main_df
```

```python
main_df.fillna(0)
```

```python
main_df.dtypes
```

```python
# Convert int64 data types to float64

main_df['avgDeathsPerYear'] = main_df['avgDeathsPerYear'].astype(float)
main_df['medIncome'] = main_df['medIncome'].astype(float)
main_df['popEst2015'] = main_df['popEst2015'].astype(float)
main_df.dtypes
```

```python
# Assign X (data) and y (target)

X = main_df[['avgAnnCount', 'avgDeathsPerYear', 'incidenceRate']]
y = main_df["TARGET_deathRate"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
import statsmodels.api as sm

mod_wls = sm.WLS(y, X, weights=1./3)
res_wls = mod_wls.fit()
```

```python
print(res_wls.summary())
```

```python
from sklearn.metrics import mean_squared_error
import math

y_pred = res_wls.predict(X)
 
MSE = mean_squared_error(y, y_pred)
 
RMSE = math.sqrt(MSE)

print("Root Mean Square Error:\n")
print(RMSE)
```

```python
from scipy import stats

def normality_of_residuals_test(res_wls):
    '''
    Function for running 4 statistical tests to 
    investigate the normality of residuals.
    
    Arg:
    * model - fitted OLS models from statsmodels
    '''
    jb = stats.jarque_bera(res_wls.resid)
    sw = stats.shapiro(res_wls.resid)
    ad = stats.anderson(res_wls.resid, dist='norm')
    ks = stats.kstest(res_wls.resid, 'norm')
    
    print(f'Jarque-Bera test ---- statistic: {jb[0]:.4f}, p-value: {jb[1]}')
    print(f'Shapiro-Wilk test ---- statistic: {sw[0]:.4f}, p-value: {sw[1]:.4f}')
    print(f'Kolmogorov-Smirnov test ---- statistic: {ks.statistic:.4f}, p-value: {ks.pvalue:.4f}')
    print(f'Anderson-Darling test ---- statistic: {ad.statistic:.4f}, 5% critical value: {ad.critical_values[2]:.4f}')
    print('If the returned AD statistic is larger than the critical value, then for the 5% significance level, the null hypothesis that the data come from the Normal distribution should be rejected. ')
    
normality_of_residuals_test(res_wls)
```

```python
import seaborn as sn
from matplotlib.pyplot import figure

figure(figsize=(16, 16), dpi=80)

corr_df = X.corr(method='pearson')
print("-----------------------------------------------Create A Correlation Plot---------------------------------------------------")

mask = np.zeros_like(corr_df)
mask[np.triu_indices_from(mask)] = True

sn.heatmap(corr_df, cmap='RdYlGn_r', vmax=1.0, vmin=-1.0, mask = mask, linewidths=2.5, annot=True)

plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```
